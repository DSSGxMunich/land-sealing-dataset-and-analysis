# Document categorization 

Many of the files we downloaded are actually not building plans. The idea of these functions is to detect which of these files are building plans, based on the text contained in the PDF and the planart in the original dataframe. 

## Step 1: prepare the data. 

- Change the paths for the folders that contain the text data from the pdfs and the original dataframe downloaded from the NRW portal. 
- Identify the columns that contain the name of the file, and the text content in the text data.

```python
# specify file path
INPUT_TEXT_DF_PATH = os.path.join( "..", "data", "nrw", "text", "nrw_final_all.csv")
INPUT_ORIGINAL_DF_PATH = os.path.join( "..", "data", "nrw", 'NRW_BP_with_links.geojson')

# specify relevant column names
ID_COLUMN='filename'
TEXT_COLUMN='content'
```

## Step 2: run the keyword and building plan detector

```python
BP_detect = run_bp_keyword_detector(INPUT_TEXT_DF_PATH, INPUT_ORIGINAL_DF_PATH, ID_COLUMN, TEXT_COLUMN, sample_n=35)
```

This returns a DF with the columns
- `filename`, `document_id`, `document_category` (on document level)
- `plan_type_code__parcel`, `plan_type__parcel`, `legal_procedure_code__parcel`, `legal_procedure__parcel` (on a land parcel level)

The latter originate from the original DF and refer to each land parcel, but might be useful for the document classification ex-post. The coding of these columns is found [here](https://www.bauportal.nrw/system/files/media/document/file/21-12-08-bauleitplanung_benutzerhandbuch.pdf).
